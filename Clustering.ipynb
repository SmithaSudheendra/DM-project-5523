```python
df = pd.read_csv('US_states_policies_mobility.csv')
df.describe()
```

```python
df_sub = df[['state','4.2.2', '4.3.3','4.3.4','4.5.2',
             'retail_recreation','grocery_pharmacy','parks','transit','workplaces','residential','population_fraction_infected']]
df_sub.corr(method ='pearson')
```

```python
df_norm = df[['4.2.2', '4.3.3','4.3.4','4.5.2',
             'retail_recreation','grocery_pharmacy','parks','transit','workplaces','residential']]
for col in ['4.2.2','4.3.3','4.3.4','4.5.2']:
    df_norm[col] = (df[col]-df[col].min())/(df[col].max()-df[col].min())
    
for col in ['retail_recreation','grocery_pharmacy','parks','transit','workplaces','residential']:
    df_norm[col] = df[col]/100.0
```

```python
df_norm.corr(method ='pearson')
```

```python
df_norm.describe()
```

# Dimensionality Reduction with t-SNE
Using t-SNE we can reduce our high dimensional features vector into 2 dimensional plane. In the process, t-SNE will keep similar instances together while trying to push different instances far from each other. Resulting 2-D plane can be useful to see which articles cluster near each other:

# Overall

```python
from sklearn.manifold import TSNE

data = df_norm[['retail_recreation','grocery_pharmacy','transit','workplaces']]
tsne = TSNE(verbose=1, perplexity=5)
X_embedded = tsne.fit_transform(data)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

# sns settings
sns.set(rc={'figure.figsize':(5,5)})

# colors
palette = sns.color_palette("bright", 1)

# plot
sns.scatterplot(X_embedded[:,0], X_embedded[:,1], palette=palette)

plt.title("Covid-19 spread due to mobility in the US")
# plt.savefig("plots/t-sne_covid19.png")
plt.show()
```

```python
plt.figure(figsize=(10,8)) 
sns.heatmap(df_sub.corr(), annot=True, cmap='cubehelix_r')
plt.show()
```

```python
from sklearn.cluster import KMeans

k = 2
kmeans = KMeans(n_clusters=k, n_jobs=-1, verbose=10)
y_pred = kmeans.fit_predict(data)

# sns settings
sns.set(rc={'figure.figsize':(5,5)})

# colors
palette = sns.color_palette("bright", len(set(y_pred)))

# plot
sns.scatterplot(X_embedded[:,0], X_embedded[:,1], hue=y_pred, legend='full', palette=palette)
plt.title("t-SNE States - Clustered")
# plt.savefig("plots/t-sne_covid19_label.png")
plt.show()
```

```python
x = list(df["population_fraction_infected"])
plt.scatter(df["population_fraction_infected"], y_pred)
plt.xlabel('data points')
plt.ylabel('Output variable')
```

```python
class_0 = []
class_1 = []

for i in range(len(x)):
    if y_pred[i] == 1:
        class_1.append(x[i])
    else:
        class_0.append(x[i])
        
import numpy as np
print('Class_0 values:', class_0,'\n')
print('Mean Class_0:', np.mean(class_0), '\n')
print('Std Class_0:', np.std(class_0), '\n')

print('Class_1 values:', class_1,'\n')
print('Mean Class_1:', np.mean(class_1), '\n')
print('Std Class_1:', np.std(class_1), '\n')
```

# Spectral Clustering

```python
from sklearn.cluster import SpectralClustering
from sklearn.neighbors import kneighbors_graph
```

```python
connectivity = kneighbors_graph(data, 4, include_self=False)
connectivity = 0.5 * (connectivity + connectivity.T)

k=2
#ward = cluster.AgglomerativeClustering(n_clusters=params['n_clusters'], linkage='ward', connectivity=connectivity)
spectral = SpectralClustering(n_clusters=k, eigen_solver='arpack', affinity="nearest_neighbors")
y_fit = spectral.fit(data)
y_pred = y_fit.labels_

# sns settings
sns.set(rc={'figure.figsize':(5,5)})

# colors
palette = sns.color_palette("bright", len(set(y_pred)))

# plot
sns.scatterplot(X_embedded[:,0], X_embedded[:,1], hue=y_pred, legend='full', palette=palette)
plt.title("t-SNE Covid-19 Articles - Clustered")
# plt.savefig("plots/t-sne_covid19_label.png")
plt.show()
```

```python
x = list(df["population_fraction_infected"])
plt.scatter(df["population_fraction_infected"], y_pred)
plt.xlabel('population fraction infected')
plt.ylabel('clusters')
```

```python
class_0 = []
class_1 = []

for i in range(len(x)):
    if y_pred[i] == 1:
        class_1.append(x[i])
    else:
        class_0.append(x[i])

import numpy as np
print('Class_0 values:', class_0,'\n')
print('Mean Class_0:', np.mean(class_0), '\n')

print('Class_1 values:', class_1,'\n')
print('Mean Class_1:', np.mean(class_1), '\n')
```
